```python
from tkinter import *

class Calci:
    def __init__(self,win):
        #Expression for evaluation
        self.exp=""
        
        #Heading
        self.lbl=Label(win,text="CALCULATOR",bg="tan4",fg="tomato",font=("cooper black",20,"bold"))
        self.lbl.place(x=28,y=10)
        
        #Display
        self.display=Entry(win,bd=4,bg="snow",fg="blue",font=("arial",12,"bold"),state="readonly")
        self.display.place(x=41,y=50)
        
        #Buttons for calculator
        self.btn1=Button(win,text="1",bg="gray",width=4,height=1,command=self.btn1_hnd,font=("calisto mt",12,"bold"))
        self.btn2=Button(win,text="2",bg="gray",width=4,height=1,command=self.btn2_hnd,font=("calisto mt",12,"bold"))
        self.btn3=Button(win,text="3",bg="gray",width=4,height=1,command=self.btn3_hnd,font=("calisto mt",12,"bold"))
        self.btn4=Button(win,text="4",bg="gray",width=4,height=1,command=self.btn4_hnd,font=("calisto mt",12,"bold"))
        self.btn5=Button(win,text="5",bg="gray",width=4,height=1,command=self.btn5_hnd,font=("calisto mt",12,"bold"))
        self.btn6=Button(win,text="6",bg="gray",width=4,height=1,command=self.btn6_hnd,font=("calisto mt",12,"bold"))
        self.btn7=Button(win,text="7",bg="gray",width=4,height=1,command=self.btn7_hnd,font=("calisto mt",12,"bold"))
        self.btn8=Button(win,text="8",bg="gray",width=4,height=1,command=self.btn8_hnd,font=("calisto mt",12,"bold"))
        self.btn9=Button(win,text="9",bg="gray",width=4,height=1,command=self.btn9_hnd,font=("calisto mt",12,"bold"))
        self.btn0=Button(win,text="0",bg="gray",width=4,height=1,command=self.btn0_hnd,font=("calisto mt",12,"bold"))
        self.btnadd=Button(win,text="+",bg="dark slate gray",width=4,height=1,command=self.btnadd_hnd,font=("calisto mt",14,"bold"))
        self.btnsub=Button(win,text="-",bg="dark slate gray",width=4,height=1,command=self.btnsub_hnd,font=("calisto mt",14,"bold"))
        self.btnmul=Button(win,text="x",bg="dark slate gray",width=4,height=1,command=self.btnmul_hnd,font=("calisto mt",14,"bold"))
        self.btndiv=Button(win,text="/",bg="dark slate gray",width=4,height=1,command=self.btndiv_hnd,font=("calisto mt",14,"bold"))
        self.btnequ=Button(win,text="=",bg="green",width=4,height=1,command=self.btnequ_hnd,font=("calisto mt",12,"bold"))
        self.btnclr=Button(win,text="C",bg="red",width=4,height=1,command=self.btnclr_hnd,font=("calisto mt",12,"bold"))
        self.btnpnt=Button(win,text=".",bg="gray",width=4,height=1,command=self.btnpnt_hnd,font=("calisto mt",12,"bold"))
        self.btnbl=Button(win,text="(",bg="gray",width=4,height=1,command=self.btnbl_hnd,font=("calisto mt",12,"bold"))
        self.btnbr=Button(win,text=")",bg="gray",width=4,height=1,command=self.btnbr_hnd,font=("calisto mt",12,"bold"))
        self.btnper=Button(win,text="%",bg="dark slate gray",width=4,height=1,command=self.btnper_hnd,font=("calisto mt",14,"bold"))
        self.lbx=Listbox(window,width=27,height=2,bd=2,bg="brown4",fg="white",font=("arial",12,"bold"))
        
        #Footer of calculator
        self.lbl1=Label(win,text="Developed by SHRUTI",bg="tan4",fg="snow2",font=("arial rounded mt bold",10))
        self.lbl1.place(x=120,y=420)
        
        #positioning the buttons and labels
        self.btn1.place(x=20,y=100)
        self.btn2.place(x=80,y=100)
        self.btn3.place(x=140,y=100)
        self.btn4.place(x=20,y=150)
        self.btn5.place(x=80,y=150)
        self.btn6.place(x=140,y=150)
        self.btn7.place(x=20,y=200)
        self.btn8.place(x=80,y=200)
        self.btn9.place(x=140,y=200)
        self.btn0.place(x=80,y=250)
        self.btnadd.place(x=200,y=98)
        self.btnsub.place(x=200,y=148)
        self.btnmul.place(x=200,y=198)
        self.btndiv.place(x=200,y=248)
        self.btnequ.place(x=140,y=300)
        self.btnclr.place(x=20,y=300)
        self.btnpnt.place(x=80,y=300)
        self.btnbl.place(x=20,y=250)
        self.btnbr.place(x=140,y=250)
        self.btnper.place(x=200,y=298)
        self.lbx.place(x=20,y=370)
        
    #Creating button handlers
    def btn1_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "1")
        self.exp=self.exp+"1"
        self.display.config(state="readonly")
    def btn2_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "2")
        self.exp=self.exp+"2"
        self.display.config(state="readonly")
    def btn3_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "3")
        self.exp=self.exp+"3"
        self.display.config(state="readonly")
    def btn4_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "4")
        self.exp=self.exp+"4"
        self.display.config(state="readonly")
    def btn5_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "5")
        self.exp=self.exp+"5"
        self.display.config(state="readonly")
    def btn6_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "6")
        self.exp=self.exp+"6"
        self.display.config(state="readonly")
    def btn7_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "7")
        self.exp=self.exp+"7"
        self.display.config(state="readonly")
    def btn8_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "8")
        self.exp=self.exp+"8"
        self.display.config(state="readonly")
    def btn9_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "9")
        self.exp=self.exp+"9"
        self.display.config(state="readonly")
    def btn0_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "0")
        self.exp=self.exp+"0"
        self.display.config(state="readonly")
        
    def btnadd_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "+")
        self.exp=self.exp+"+"
        self.display.config(state="readonly")
    def btnsub_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "-")
        self.exp=self.exp+"-" 
        self.display.config(state="readonly")
    def btnmul_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "*")
        self.exp=self.exp+"*"
        self.display.config(state="readonly")
    def btndiv_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "/")
        self.exp=self.exp+"/" 
        self.display.config(state="readonly")
    def btnpnt_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, ".")
        self.exp=self.exp+"." 
        self.display.config(state="readonly")
    def btnbl_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "(")
        self.exp=self.exp+"(" 
        self.display.config(state="readonly")
    def btnbr_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, ")")
        self.exp=self.exp+")" 
        self.display.config(state="readonly")
    def btnper_hnd(self):
        self.display.config(state="normal")
        self.display.insert(END, "%")
        self.exp=self.exp+"%"  
        self.display.config(state="readonly")
    
    #displaying result and handling exception
    def btnequ_hnd(self):
        try:
            self.display.config(state="normal")
            self.res=eval(self.exp)
            self.display.delete(0,"end")
            self.display.insert(END,float(self.res))
            self.lbx.insert(END,self.res)
            self.exp=str(self.res)
            self.display.config(state="readonly")
        except Exception as e:
            self.display.config(state="normal")
            self.display.delete(0,"end")
            self.display.insert(END,"INVALID!! Press C")
            self.display.config(state="readonly")
    def btnclr_hnd(self):
        self.display.config(state="normal")
        self.display.delete(0,"end")
        self.exp=""
        self.display.config(state="readonly")
 
#Creating window object
window=Tk()
window.title("Python Calculator")
cal=Calci(window)
    
window.configure(bg="tan4")
window.geometry("280x450+800+100")
window.resizable(False,False)
window.mainloop()
```
